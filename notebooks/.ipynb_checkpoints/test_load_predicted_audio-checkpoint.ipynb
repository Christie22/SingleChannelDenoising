# Test predicted audio
This script load a predicted audio file, i.e. the output of `denoise.py`.   
It allows the user to listen to it, plot the spectrogram, and perform various analyses (calculate metrics, etc).

## Caveats
Currently a work in progress! :D

```python
import librosa as lr
import librosa.display as lrd
import numpy as np
import IPython.display as ipd
```

```python
n_fft = 512
win = n_fft
hop = win // 4

```

```python
### LOAD ORIGINAL ###


x, sr = lr.load('/data/riccardo_datasets/npr_news/train/newscast220831.wav', sr=None)

trange = slice(0, sr*10)

s = lr.stft(x[trange], n_fft=n_fft, hop_length=hop, win_length=win)
lrd.specshow(lr.amplitude_to_db(np.abs(s)))
ipd.Audio(x[trange], rate=sr)

```

```python
### LOAD PREDICTED ###

y3, sr = lr.load('/data/riccardo_datasets/npr_news/outputs/9.wav', sr=None)

s = lr.stft(y3[trange], n_fft=n_fft, hop_length=hop, win_length=win)
lrd.specshow(lr.amplitude_to_db(np.abs(s)))
ipd.Audio(y3[trange], rate=sr)
```
